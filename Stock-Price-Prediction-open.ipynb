```python
df_train=pd.read_csv('../input/google-stock-price/Google_Stock_Price_Train.csv')
df_test= pd.read_csv('../input/google-stock-price/Google_Stock_Price_Test.csv')
```

```python
df_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
X_temp= scaler.fit_transform(df_train[['Open']])
```

```python
X_temp
```

```python
X=[]
for i in range(0,1198):
    temp=X_temp[i:i+61,0]
    X.append(temp)
```

```python
X=np.array(X)
```

```python
y=X[:,-1]
```

```python
X=X[:,:60]
```

```python
X.shape
```

```python
X=X.reshape(-1,60,1)
```

```python
X.shape
```

```python
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.models import Sequential
```

```python
model= Sequential()
model.add(LSTM(50, return_sequences= True, input_shape=(60,1)))
model.add(LSTM(50, return_sequences= True))
model.add(LSTM(50, return_sequences= True))
model.add(LSTM(50, return_sequences= False))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X,y,epochs=50, batch_size=32)
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.plot(hist.history['loss'])
```

```python
y_pred= model.predict(X)
```

```python
y_pred= scaler.inverse_transform(y_pred)
```

```python
tr= df_train['Open'][60:].reset_index(drop= True)
```

```python
plt.plot(tr, c='blue')
plt.plot(y_pred, c='red')
```
